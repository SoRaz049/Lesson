```python
years = [2000 + x for x in range(25)]

# Generate a single random integer between 50 and 70
weight = [int(np.random.uniform(50,70)) for _ in range (25)]
```

```python
plt.figure(figsize=(12,4))
plt.plot(years, weight, c= "green", lw= 3, linestyle = "--")
plt.xlabel("Years")
plt.ylabel("Weight")
plt.title("Weight Chart")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.scatter(years, weight, c= "green", lw= 3, linestyle = "--")
plt.xlabel("Years")
plt.ylabel("Weight")
plt.title("Weight Chart")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.bar(years, weight, color = "grey")
plt.xlabel("Years")
plt.ylabel("Weight")
plt.title("Weight Chart")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.hist(years)
plt.xlabel("Years")
plt.ylabel("Weight")
plt.title("Weight Chart")
plt.show()
```
