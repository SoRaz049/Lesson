```python
years = [2000 + x for x in range(25)]

# Generate a single random integer between 50 and 70
weight = [int(np.random.uniform(50,70)) for _ in range (25)]
```

```python
plt.figure(figsize=(12,4))
plt.plot(years, weight, c= "green", lw= 3, linestyle = "--")
plt.xlabel("Years")
plt.ylabel("Weight")
plt.title("Weight Chart")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.scatter(years, weight, c= "green", lw= 3, linestyle = "--")
plt.xlabel("Years")
plt.ylabel("Weight")
plt.title("Weight Chart")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.bar(years, weight, color = "grey")
plt.xlabel("Years")
plt.ylabel("Weight")
plt.title("Weight Chart")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.hist(years, bins= 15, cumulative= True)
plt.xlabel("Years")
plt.ylabel("Weight")
plt.title("Weight Chart")
plt.show()
```

```python
import seaborn as sns
sns.histplot(years,  kde = True)
plt.title("Weight Chart")
plt.show()
```

```python
plt.boxplot(weight)
plt.xlabel("Years")
plt.ylabel("Weight")
plt.title("Box Chart")
plt.show()
```

```python

years_income = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
income = [55, 56, 62, 61, 72, 72, 73, 75, 69, 65, 77]
income_ticks = list(range(50, 81, 2))

plt.figure(figsize=(12, 4))
plt.plot(years_income, income, color="#f65077", marker='o', linestyle='-') 
plt.scatter(years_income[0], income[0], color='red', marker='*', s=500)
plt.xlabel("Years")
plt.ylabel("Income in USD")
plt.title("Annual income in USD")
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks])
plt.legend()
plt.show()

```

```python
stock_a = [186, 187, 99, 181, 161, 100, 102,110]
stock_b = [96, 95, 102, 104, 105, 103, 109,135]
stock_c = [110, 115, 100, 105, 190, 98, 95,85]
years_stock = [2000, 2003, 2006, 2009, 2012, 2015, 2018, 2021]

plt.figure(figsize=(10, 10))
plt.plot(years_stock, stock_a, color = 'red', marker = '4', label = "Microsoft")
plt.plot(years_stock, stock_b, color = 'blue', marker = 's', label = "Apple")
plt.plot(years_stock, stock_c, color = 'green', marker = '^', label = "Nokia")
plt.legend()
plt.show()
```

```python
x= np.arange(100)

fig, axs = plt.subplots(2,2 , figsize=(12,4))


axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("Sine Wave")

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title("Cosine Wave")

axs[1,0].plot(x, np.tan(x))
axs[1,0].set_title("Tan Wave")

axs[1,1].plot(x, np.log(x))
axs[1,1].set_title("Log Wave")


plt.tight_layout()
plt.show()
```

```python
ax = plt.axes(projection = "3d")

x = np.random.random(100)
y = np.sin(x)
z = np.cos(x)

ax.plot(x,y,z)
ax.set_title("3D plot")
plt.show()
```

```python
ax = plt.axes(projection = "3d")
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

X, Y = np.meshgrid(x,y)
Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X,Y,Z, cmap = "Spectral")
ax.set_title("3D plot")
plt.show()

```
